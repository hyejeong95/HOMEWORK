## Facets
-split your plot into facets.subplots that each display one subset of the data. 


```r
ggplot(data=mpg)+
geom_point(mapping=aes(x=displ,y=hwy))+
facet_wrap(~class,nrow=2)
#facet_wrap() 는 그룹별로 산포도룰 분리해서 그려보자. 
# ~ : followed by a variable name

```

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy))+facet_grid(drv ~ cyl)

#facet_grid() : 함수 안의 ~ 좌/우 변수를 각각 행/열로 나누어 2차원으로 
#sub그래프들을 그려준다. 
```

```r
ggplot(data=mpg)+
geom_line(mapping=aes(x=displ,y=hwy))  #모든 점들을 하나하나 다 연결해준다. 그러므로 정확한 흐름이 보이지 않는다. 

ggplot(data=mpg)+
geom_smooth(mapping=aes(x=displ,y=hwy)) 
```

```r
ggplot(data=mpg)+geom_smooth(mapping=aes(x=displ,y=hwy,group=drv))
#group aesthetic to a categorical variable to draw multiple objects. 
```

```r
ggplot(data=mpg)+geom_smooth(mapping=aes(x=displ,y=hwy,color=drv),show.legend=FALSE)
```

```r
ggplot(data=mpg,mapping=aes(x=displ,y=hwy))+
geom_point(mapping=aes(color=class))+geom_smooth()

#aes안에는 표의 속성을 넣어줘야 한다. 
```

```r
ggplot(data=mpg,mapping=aes(x=displ,y=hwy))+
geom_point()+geom_smooth()
```

```r
ggplot(data=mpg)+geom_point(mapping=aes(x=displ,y=hwy))
+theme_bw()+ theme(axis.line=element_line(size=.8,colour="black"),
                   panel.grid.minor=element_blank(),
                   panel.broder=element_blank(),
                   text=element_text(size=25))
```

## arrange() 
reorders the rows according to single or multiple 

```r
arrange(flights,year,month,day)
```

```r
#des() : re-order by a column in descending order
arrange(flights,desc(dep_delay))
```

## select()
return a subset of the columns 

```r
select(flights,year,month,day)
```

```r
select(flights,-(year:day))
```

## mutate()
add columns from existing data

```r
flights_sml<-select(flights,year:day,ends_with("delay"),distance,air_time)
```

```r
flights_sml  #ends_with():()안에 들어가는 말로 끝나는 모든 변수들을 선택하겠다. 
```

```r
mutate(flights_sml,gain=dep_delay-arr_delay)  #gain이라는 변수를 추가해준다. 
```

```r
#transmute()라고 하면 새로 추가하는 변수를 유지할 수 있다. 
```

## Summarise()
group_by을 이용하여 준다.

```r
by_day<-group_by(flights,year,month,day)
summarise(by_day,delay=mean(dep_delay,na.rm=TRUE))
```

## Pipe operators
%>% 사용 

```r
by_dest<-group_by(flights,dest)
```

```r
print(by_dest)
delay<-summarise(by_dest,count=n(),dist=mean(distance,na.rm=TRUE),delay=mean(arr_delay,na.rm=TRUE))
```

```r
delay<-filter(delay,count>20,dest!="HNL")
delay
```
